```python
#INTRODUCTORY QUESTION 4:-
df['population']=(df['cases']/df['cases100k'])*100000#Column created
df
```

```python
#TASK QUESTION 1:-
#Find the columns having null values
df.isnull().sum()
```

```python
#Displaying the rows having null values
nan_val=df.isnull()
rowNAN=nan_val.any(axis=1)
rowswithNAN=df[rowNAN]
rowswithNAN
```

```python
#Filling null values with mean as it appropriate with the numerical categorical data with less null values.
#Done with the application of python dictionary
df.fillna({'cases':df['cases'].mean(),'cases100k':df['cases100k'].mean(),'cases7days':df['cases7days'].mean(),
           'cases7days100k':df['cases7days100k'].mean(),'deaths':df['deaths'].mean(),
           'deaths100k':df['deaths100k'].mean(),'deaths7days':df['deaths7days'].mean(),'deaths7days100k':df['deaths7days100k'].mean(),
           'fatality_rate':df['fatality_rate'].mean(),'population':df['population'].mean()}, inplace=True)
```
